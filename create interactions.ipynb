```python
import sqlite3

con = sqlite3.connect('database.sqlite3')
cur = con.cursor()
print("query starting")
cur.execute("SELECT compounds.compound_id, compounds.standard_inchi, proteins.protein_id, proteins.protein_sequence, t.standard_value FROM proteins INNER JOIN t on protein_id = target_id INNER JOIN compounds on t.compound_id = compounds.compound_id;")
interaction_data = cur.fetchall()
print("query complete")
cur.execute("DROP TABLE IF EXISTS interactions")
cur.execute("CREATE TABLE IF NOT EXISTS interactions (compound_id, standard_inchi, target_id, target_sequence, standard_value);") # use your column names here
count = 0
for i in interaction_data:
    if i[3] != "ASERROR":
        cur.execute("INSERT INTO interactions (compound_id, standard_inchi, target_id, target_sequence, standard_value) VALUES (?, ?, ?, ?, ?);", i)
    else:
        count+=1
        print("count")
con.commit()
con.close()
```
